## For which of the models already studied is the curse of dimensionality relevant and explain why.

Проклятие размерности сильно влияет на любые методы, основанные на расстояниях:
- KNN

Почему:

С увеличением размерности пространства происходят следующие эффекты:

- Разреженность данных: Объём пространства растёт экспоненциально, а количество данных — нет → точки становятся далеко друг от друга.

- Потеря смысла расстояний: Разница между ближайшим и самым дальним соседом становится незначительной:

- Переобучение: Модель легко подгоняется под шум, так как признаков слишком много.

- Рост вычислительной сложности: обработка и хранение становятся дорогими.


Средне оно влияет на линейные модели: из-за множества параметров легко может возникнуть переобучение, но его можно контролировать через регуляризацию.

Деревья решений можно сказать, что почти не подвержены: не используют расстояния, делают разбиения по одному признаку.

## What is the difference between PCA and SVD?

PCA (Principal Component Analysis)
- Метод снижения размерности
- Ищет направления максимальной дисперсии данных
- Работает с ковариационной матрицей

SVD (Singular Value Decomposition)
- Метод факторизации с матричным разложением на ортогональные матрицы и диаганальную матрицу с сингулярными числами
- Работает с любой матрицей

| PCA | SVD |
| - | - |
| Алгоритм анализа данных | Матричное разложение |
| Требует центрирования | Не требует |
| Работает через ковариацию | Работает напрямую с матрицей |
| Специализирован под снижение размерности | Универсальный инструмент |


## What is the difference between NMF and SVD?

| Свойство | SVD | NMF |
| - | - | - |
| Знаки | любые | только ≥ 0 |
| Интерпретируемость | низкая | высокая |
| Ортогональность | да | нет |
| Применение | общее | темы, изображения, рекомендации |

Интуиция
- SVD "математически оптимальное сжатие"
- NMF "сборка из частей"

Пример:
- SVD: лицо как комбинация абстрактных паттернов
- NMF: лицо как сумма глаз + носа + рта

## Describe the structure of the Locally Linear Embedding dimension reduction algorithm.

LLE — это метод нелинейного снижения размерности, который сохраняет **локальную геометрию данных**.

---

### Идея

Каждая точка представляется как линейная комбинация своих соседей.
Эта структура сохраняется при переходе в пространство меньшей размерности.

---

### Шаги алгоритма

### Шаг 1: Поиск соседей

Для каждой точки (x_i):

- находим (k) ближайших соседей

---

### Шаг 2: Восстановление точки через соседей

Находим веса (w_{ij}), минимизируя:

$
| x_i - \sum_j w_{ij} x_j |^2
$

при условии:

$
\sum_j w_{ij} = 1
$

Это означает, что каждая точка выражается через своих соседей.

---

### Шаг 3: Построение нового пространства

Ищем низкоразмерные представления (y_i), минимизируя:

$
| y_i - \sum_j w_{ij} y_j |^2
$

Сохраняем те же веса и сохраняем локальную структуру.

---

### Шаг 4: Решение задачи

Задача сводится к нахождению собственных векторов матрицы.

---

### Свойства LLE

+ Сохраняет локальную геометрию
+ Хорош для "скрученных" многообразий
- Чувствителен к шуму
- Требует выбора числа соседей
